```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
```

```python
df_train = pd.read_csv('../data/input/google-quest-challenge/train.csv')
```

```python
def build_vocab(texts: pd.Series):
    sentences = texts.progress_apply(lambda x: x.split()).to_numpy()
    vocab = {}
    for sentence in sentences:
        for word in sentence:
            try:
                vocab[word] += 1
            except KeyError:
                vocab[word] = 1
    return vocab

vocab = build_vocab(df_train['question_body'])
len(vocab)
```

```python
from nltk.tokenize.treebank import TreebankWordTokenizer

tree_tokenizer = TreebankWordTokenizer()
```
